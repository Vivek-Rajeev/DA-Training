```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.DataFrame({
    "Order_time":[7,8,7,9,7],
    "Delivery_time":[8.10,8.30,7.40,10.00,7.40],
    "Delivery_partner":['Raga','Ramu','Sudha','somu','rakesh'],
    "Customer_ratings":[4,5,4,3,5]
})

print(df)

df['processing']=df['Delivery_time']-df['Order_time']
print("Processing Time",df['processing'])

avg_delivery_time=df['processing'].mean()
print(f"Average delivery time: {avg_delivery_time}")

peak_order_hours=df['Order_time'].value_counts().idxmax()
print(f"Peak order hours: {peak_order_hours}")

most_delayed_orders=df[df['processing'].max()==df['processing']]
print("Most delayed orders:")
print(most_delayed_orders)

top_ratings=df['Customer_ratings'].max()

top_rated_delivery_partner=df[df['Customer_ratings']==top_ratings]['Delivery_partner'].tolist()
print(f"Top rated delivery partner:{top_rated_delivery_partner} by rating  {top_ratings}")


plt.figure(figsize=(8,5))
plt.bar(df['Delivery_partner'],df['processing'])
plt.xlabel('Delivery Partner')
plt.ylabel('Processing Time')
plt.title('Processing Time by Delivery Partner')
plt.show()

plt.figure(figsize=(8,5))
plt.bar(df['Delivery_partner'],df['Customer_ratings'])
plt.xlabel('Delivery Partner')
plt.ylabel('Customer Ratings')
plt.title('Customer Ratings by Delivery Partner')
plt.show()

plt.pie(df['Customer_ratings'],labels=df['Delivery_partner'])
plt.title('Customer Ratings')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.DataFrame({
    "product_purchases":['peanut butter','chocolate','shoes','earbuds','mobile','laptop'],
    "products_sold":[40,30,50,70,20,10],
    "cart_abandonment":[0,1,1,0,1,0],
    "customer_ratings":[4,5,4,3,5,4],
    "payment_method":['upi','cash','cash','upi','cash','cash']
})

print(df)

most_sold_product = df[df['products_sold'].max()==df['products_sold']]
print("Most sold product:")
print(most_sold_product)

most_used_payment_method=df['payment_method'].value_counts().idxmax()
print(f"Most used payment method: {most_used_payment_method}")

poor_product_ratings=df[df['customer_ratings'].min()==df['customer_ratings']]
print("Poor product ratings:")
print(poor_product_ratings)

high_revenue_products=df[df['products_sold']>30]
print("High revenue products:")
print(high_revenue_products)

plt.figure(figsize=(8,5))
plt.bar(df['product_purchases'],df['products_sold'])
plt.xlabel('Product')
plt.ylabel('Products Sold')
plt.title('ecommerence analysis')
plt.show()

plt.figure(figsize=(8,5))
plt.pie(df['products_sold'],labels=df['product_purchases'])
plt.title('Products Sold')
plt.show()
```

```python
from numpy import astype
import pandas as pd
import matplotlib.pyplot as plt

df=pd.DataFrame({
    "Transaction amount":[10000.07,20000,5000.05,75000.02,40000,36000],
    "location":['bng','hyd','vizag','tirupathi','chennai','mumbai'],
    "time":[1.20,12.30,5.30,4.50,6.20,3.30],
    "account activity":['yes','no','yes','no','yes','no']
})

print(df)
unusual_transactions = df[
    df['Transaction amount'] != df['Transaction amount'].astype(int)
]

print("Unusual transactions:")
print(unusual_transactions)

high_risk_amount=df[df['Transaction amount']>50000]
print("High risk transactions:")
print(high_risk_amount)

fraud_pattrens = df[df['account activity'] == 'no']
print("Fraud patterns:")
print(fraud_pattrens)

plt.figure(figsize=(8,5))
plt.bar(df['location'],df['Transaction amount'])
plt.xlabel('location')
plt.ylabel('Transaction amount')
plt.title('Transaction amount by location')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.DataFrame({
    "daily sales":[400,300,250,500,150],
    "product stock":[500,300,200,500,300],
    "expiry dates":[50,100,90,30,60],
    "seasonal demand":[1,1,0,1,0]
})

print(df)

avg_sales=df['daily sales'].mean()
print(f"Average sales: {avg_sales}")

fast_moving_products=df[
    (df['seasonal demand']==1 )& (df['daily sales']>df['daily sales'].mean())]
print("Fast moving products:")
print(fast_moving_products)

slow_moving_products=df[
    (df['seasonal demand']==0 )& (df['daily sales']<df['daily sales'].mean())]
print("Slow moving products:")
print(slow_moving_products)

df['stock_remaining']=df['product stock']-df['daily sales']
print("Stock remaining:")
print(df[['stock_remaining','daily sales']])

out_of_stock_products=df[df['product stock']<=df['daily sales']]
print("Out of stock products:")
print(out_of_stock_products)
```
