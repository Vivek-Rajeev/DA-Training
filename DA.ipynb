```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

marks = [95,85,68,73,94]
hours = [7,8,7,9,6]

print("Mean:",np.mean(marks))
print("Median:",np.median(marks))
print("Mode:",pd.Series(marks).mode()[0])
print("Standard Deviation:",np.std(marks))
print("Variance:",np.var(marks))
print("Covariance:",np.cov(marks,hours))
print("Correlation:",np.corrcoef(marks,hours))

plt.figure(figsize=(6,5))
plt.scatter(hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()

plt.figure(figsize=(8,5))

students = ['S1','S2','S3','S4','S5']

plt.bar(students, marks)

plt.title("Student Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()

```

```python
import statistics
import matplotlib.pyplot as plt

mem_names = ["amara","riya","kavin","neha","raju"]

ages=[22,23,22,25,24]

workout_hours=[2,1.5,2.5,1.3,2]
calories_burned=[500,300,450,600,700]
print("Fitness Center")
for i in range(len(mem_names)):
  print(mem_names[i],
        ages[i],
        workout_hours[i],
        calories_burned[i])

average_workout=sum(workout_hours)/len(workout_hours)
print("Average workout hours:",average_workout)

average_calories=sum(calories_burned)/len(calories_burned)
print("Average calories burned:",average_calories)

max_calories=max(calories_burned)
print("Maximum calories burned:",max_calories)

min_calories=min(calories_burned)
print("Minimum calories burned:",min_calories)

median_calories=statistics.median(calories_burned)
print("Total calories burned:",median_calories)

std_dev=statistics.stdev(calories_burned)
print("Standard Deviation=",std_dev)

index = calories_burned.index(max_calories)
print("Most Active Member=",mem_names[index])

index2 = calories_burned.index(min_calories)
print("Most InActive Member=",mem_names[index2])

plt.bar(mem_names,calories_burned)
plt.xlabel("members")
plt.ylabel("calories burned")
plt.title("Fitness Center")
plt.show()

plt.scatter(mem_names,calories_burned)
plt.xlabel("members")
plt.ylabel("calories burned")
plt.title("Fitness Center")
plt.show()

plt.pie(calories_burned,labels=mem_names)
plt.title("Fitness Center")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ecommerce_sales_data.csv")
print(data.head())

total_customers = 0

for i in data['CustomerID']:
  total_customers = total_customers + 1

print("Total Customers:", total_customers)

discount_issue = 0
normal_internet = 0

for discount in data['Discount']:

    if discount > 0.20:

        discount_issue = discount_issue + 1

    else:

        normal_internet = normal_internet + 1

print("Discount Diagnosis:", discount_issue)

payment_issue = 0
normal_payment = 0

for payment in data['PaymentMethod']:

    if payment == "Cash on Delivery":

        payment_issue = payment_issue + 1

    else:

        normal_payment = normal_payment + 1

print("Payment Diagnosis:", payment_issue)

delivery_issue = 0
normal_delivery = 0

for quantity in data['Quantity']:

    if quantity > 3:

        delivery_issue = delivery_issue + 1

    else:

        normal_delivery = normal_delivery + 1

print("Delivery Diagnosis:", delivery_issue)

if discount_issue > payment_issue and discount_issue > delivery_issue:

    root_cause = "Discount Diagnosis"

elif payment_issue > discount_issue and payment_issue > delivery_issue:

    root_cause = "Payment Diagnosis"

else:

    root_cause = "Delivery Diagnosis"

print("Main Root Cause:", root_cause)


labels = ['Discount Issues', 'Normal']
values = [discount_issue, normal_internet]

plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Discount Diagnosis")
plt.show()


labels = ['Payment Issues', 'Normal']
values = [payment_issue, normal_payment]

plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Payment Diagnosis")
plt.show()


labels = ['Delivery Issues', 'Normal']
values = [delivery_issue, normal_delivery]

plt.figure(figsize=(5,5))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Delivery Diagnosis")
plt.show()

labels=['Discount','Payment','Delivery']
values=[discount_issue,payment_issue,delivery_issue]

plt.figure(figsize=(5,5))
plt.pie(values,labels=labels)
plt.title("Root cause Analysis")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]
consumption = [150,100,120,100,110,200,150]
data={'days':days,'consumption':consumption}
print(data)
df=pd.DataFrame(data)
print(df)

average_consumption = df['consumption'].mean()
print("Average Water Consumption:", average_consumption)

total_water_usage = df['consumption'].sum()
print(f"Total Water Usage in the given period: {total_water_usage} litres")

tank_size = 220
print(f"Tank size: {tank_size} litres")

if total_water_usage > tank_size:
    print("Total water usage exceeds the tank size.")
elif total_water_usage < tank_size:
    print("Total water usage is within the tank size capacity.")
else:
    print("Total water usage is exactly the tank size.")

df['daily_increase'] = df['consumption'].diff().fillna(0)
print("\nDaily Water Consumption Increase:")
print(df[['days', 'consumption', 'daily_increase']])
print("Sum of daily consumtion of water:",df['daily_increase'].sum())
print("Maximum daily consumption of water:",df['daily_increase'].max())
print("Average:",df['daily_increase'].mean())

if((consumption[-1])+df['daily_increase'].mean())>tank_size:
  print("Total water usage exceeds the tank size.")
else:
  print("Total water usage is within the tank size capacity.NO SHORTAGES")

plt.figure(figsize=(8,5))
plt.plot(df['days'],df['consumption'])
plt.xlabel("Days")
plt.ylabel("Consumption")
plt.title("Daily Water Consumption")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

days=[1,2,3,4,5,6,7]
temp=[30,32,36,39,40,42,43]
data={'days':days,'temp':temp}
print(data)
df=pd.DataFrame(data)
print(df)

avg_temp=df['temp'].mean()
print("Average Temp:",avg_temp)

daily_increase=df['temp'].diff().fillna(0)
print("Daily Increase:",daily_increase)

sum_increase=daily_increase.sum()
print("Sum of daily increase:",sum_increase)

avg_increase=daily_increase.mean()
print("Average increase:",avg_increase)

if avg_temp > 30 and avg_increase > 0:
  print("Prediction: Weather is likely sunny.")
else:
  print("Prediction: Weather might not be sunny (or other conditions prevail).")

if(((temp[-1])+avg_increase)>avg_temp):
  print("Prediction: Weather is likely sunny.")
else:
  print("Prediction: Weather might not be sunny (or other conditions prevail).")

plt.figure(figsize=(8,5))
plt.plot(df['days'],df['temp'])
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Daily Temperature")
plt.show()
```
